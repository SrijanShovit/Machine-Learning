```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head(3)
```

```python
## dropping out unusable fields

df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head(5)
```

```python
target = df.Survived
```

```python
input = df.drop('Survived',axis='columns')
```

```python
## Converting sex to dummy variable

dummies = pd.get_dummies(input.Sex)
dummies.head(3)
```

```python
## merge dummy columns to input
input = pd.concat([input,dummies],axis=1)
input.head()
```

```python
# drop original sex column
input.drop('Sex',inplace=True,axis=1)
input.head(3)
```

```python
# checking for any NA values
input.columns[input.isna().any()]
```

```python
# taking mean for replacing NAN in age

input.Age = input.Age.fillna(input.Age.mean())
input.head(10)
```
